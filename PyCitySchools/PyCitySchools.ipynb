```python
school_results = school_data.rename(columns={"school_name": "School Name", "type": "School Type", "size": "Total Students",
                                      "budget": "Total School Budget"})

school_results = school_results.drop(['School ID'], axis=1)
```

```python
school_results = school_results.set_index('School Name')
```

```python
school_results["Per Student Budget"] = school_results['Total School Budget'] / school_results['Total Students']
```

```python
passing_read_stu = school_data_complete[school_data_complete["reading_score"]>69]
passing_math_stu = school_data_complete[school_data_complete["math_score"]>69]
passing_math_stu
```

```python
school_group = school_data_complete.groupby("school_name")
school_group
```

```python
school_group_math = passing_math_stu.groupby("school_name")
school_group_read = passing_read_stu.groupby("school_name")
```

```python
school_results["Passing Stu Math"] = school_group_math["math_score"].count()
school_results["Passing Stu Read"] = school_group_read["reading_score"].count()
```

```python
school_results["Average Math Score"] = school_group["math_score"].mean()
```

```python
school_results["Average Reading Score"] = school_group["reading_score"].mean()
school_results
```

```python
school_results["% Passing Math"] = (school_results['Passing Stu Math'] / school_results['Total Students']) * 100
school_results["% Passing Reading"] = (school_results['Passing Stu Read'] / school_results['Total Students']) * 100
school_results["% Overall Passing Rate"] = (school_results['% Passing Math'] + school_results['% Passing Reading']) / 2
```
